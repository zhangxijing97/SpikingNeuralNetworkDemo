```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
tau_m = 20.0  # membrane time constant (ms)
R = 1.0       # membrane resistance (MΩ)
V_th = -50.0  # threshold potential (mV)
V_reset = -65.0  # reset potential (mV)
V_rest = -65.0   # resting potential (mV)
I = 1.5         # input current (nA)
dt = 0.1        # time step (ms)
T = 100         # total simulation time (ms)

# Time array
time = np.arange(0, T, dt)

# Initialize membrane potential
V = np.full_like(time, V_rest)

# Simulation loop
for t in range(1, len(time)):
    dV = (-(V[t-1] - V_rest) + R * I) / tau_m * dt
    V[t] = V[t-1] + dV
    if V[t] >= V_th:
        V[t] = V_reset  # reset membrane potential

# Plotting the results
plt.plot(time, V)
plt.xlabel('Time (ms)')
plt.ylabel('Membrane Potential (mV)')
plt.title('Leaky Integrate-and-Fire (LIF) Neuron')
plt.show()
```
